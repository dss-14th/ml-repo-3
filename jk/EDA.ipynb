```python
df['urban'] = df['urban'].astype('str')
```

```python
df['urban'] = df['urban'].str.replace('1', 'Rural')
df['urban'] = df['urban'].str.replace('2', 'Suburban')
df['urban'] = df['urban'].str.replace('3', 'Urban')
```

```python
df['urban']
```

```python
# urban
df[df['urban'] == 1] = 'Rural'
df[df['urban'] == 2] = 'Suburban'
df[df['urban'] == 3] = 'Urban'
```

```python
# gender
df[df['urban'] == 1] = 'Male'
df[df['urban'] == 2] = 'Female'
df[df['urban'] == 3] = 'Other'
```

```python
# engnat
df[df['engnat'] == 1] = 'Yes'
df[df['engnat'] == 2] = 'No'
```
