## statsmodels.Logit

```python
import statsmodels.api as sm
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
```

```python
X = df_tve.drop('voted', axis=1)
y = df_tve['voted']
```

```python
cat_cols = []
for i in X.columns:
    if 'V' in i:
        cat_cols.append(i)
cat_cols
```

```python
vcl_dummy = pd.get_dummies(X[cat_cols])
```

```python
X.drop(cat_cols, axis=1, inplace=True)
```

```python
X = pd.concat([vcl_dummy, X], axis=1)
```

```python
y = lb.fit_transform(y)
```

```python
logit = sm.Logit(y, X).fit()
```

```python
logit.summary()
```

```python
logit.summary2()
```
