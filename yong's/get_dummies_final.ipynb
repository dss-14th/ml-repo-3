## 모델 적용전 범주화 전처리

```python
df = pd.read_csv('../MACH_data/data_cutoff_age18.csv')
```

```python
from sklearn.preprocessing import RobustScaler

def tran_robust(df):
    num_cols = []
    for col in df.columns:
        if ("E" in col) | ("age" in col) |("family" in col) |("elapse" in col):
            num_cols.append(col)
    rs = RobustScaler()
    df[num_cols] = rs.fit_transform(df[num_cols])
    return df
```

```python
df = tran_robust(df)
```

```python
y = df["voted"]
x = df.drop("voted", axis=1)
```

```python
x.shape, y.shape
```

### x get_dummies

```python
X = pd.get_dummies(x)
```

### y 값 라벨 인코더

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
Y = lb.fit_transform(y)
Y   # no = 0, yes = 1
```

```python
X.shape, Y.shape
```

```python
X.isnull().sum().sum()
```

```python

aa = pd.DataFrame(X.describe())
bb = aa.loc[:,((aa.loc["max",:] == 1) & (aa.loc["min",:] == 0))]
df1 = X[bb.columns].copy()

datas = []
for i in df1.columns:
    datas.append(i)
data = []
for i in datas:
    if "VCL" not in i:
        data.append(i)

not_vcl = list(set(pd.Series(data).str.split("_").str[0]))
df2 = df1[data]

for index in range(len(not_vcl)):
    
    cc = []
    
    for i,j in enumerate(list(df2.columns)):
        if not_vcl[index] in j:
            cc.append(i)
            
    
    df3 = df2.iloc[:,cc]
    df4 = pd.concat([df3,y], axis=1)
    df5 = pd.DataFrame(df4.groupby("voted").sum())
    df6 = (df5.iloc[1,:] - df5.iloc[0,:])/df5.sum() + 1
    X[df2.iloc[:,cc].columns] = (X[df2.iloc[:,cc].columns]* df6)
    
```
