#### score 컬럼

> 테스트 참여자의 마키아벨리즘에 대한 수용 수준 점수화
- 점수화 소스 데이터
    - 마키아벨리즘 test 답변 점수
    - 1점(Disagree) ~ 5점(Agree)    
>    
>    
- 점수화 공식
    - ```수용적 경향 질문지 답변 점수의 합```  +  ```불수용적 경향 질문지 답변 점수의 역변환 값 (6-점수) ```

````python
pos_col = []
neg_col = []

for col in col_list:
    if "P" in col and "Y" not in col:
        pos_col.append(col)
    if "N" in col:
        neg_col.append(col)
        
print(pos_col, '\n', '\n', neg_col)
````

````python
df["score"] = df[pos_col].sum(axis=1) + df[neg_col].apply(lambda x: 6 -x).sum(axis=1)
df["score"]
````

````python
df["score"].describe()
````

````python
## 결측치 포착..
df[df["score"]==0.0]
````

#### V, T, M 컬럼

````python
a = 6 - df.loc[:,neg_col]
b = df.loc[:,pos_col]
c = pd.concat([a,b], axis=1)
d=list(c.columns)
d
````

````python
v_score = []
t_score = []
m_score = []

for col in d:
    if "T" in col:
        t_score.append(col)
    if "M" in col:
        m_score.append(col)
    if "V" in col:
        v_score.append(col)
        
print(t_score, '\n', m_score, '\n', v_score)
````

````python
df["v_score"] = c[v_score].sum(axis=1)
````

````python
df["t_score"] = c[t_score].sum(axis=1)
````

````python
df["m_score"] = c[m_score].sum(axis=1)
````

````python
def tran_cate(df, cate, x, y):
    df[cate] = df[cate].astype('str').replace(x, y)
    return df[cate]
````

````python
cates = {
    'education' : ['Less than high school','High school','University degree','Graduate degree'],
    'urban' : ['Rural','Suburban','Urban'],
    'gender' : ['Male','Female','Other'],
    'engnat' : ['Yes','No'],
    'hand' : ['Right','Left','Both'],
    'religion' : ['Agnostic','Atheist','Buddhist','Christian(Catholic)','Christian(Mormon)','Christian(Protestant)','Christian(othrer)','Hindu','Jewish','Muslim','Sikh','Other'],
    'orientation' : ['Heterosexual','Bisexual','Homosexual','Asexual','Other'],
    'race' : ['Asian','Arab','Black','Indigenous Australian','Native American','White','Other'],
    'voted' : ['Yes','No'],
    'married' : ['Never married','Currently married','Previously married'],
}
````

````python
for x in list(cates.keys()):
    for idx, y in enumerate(cates[x]):
        if x == 'race':
            tran_cate(df, x, "{}".format((idx+1)*10) ,y)
        else:    
            tran_cate(df, x, "{}".format(idx+1) ,y)
````
